## a) Accessing data

```python
# Connect to the Hub
client = pyeodh.Client().get_catalog_service()
cmip6 = client.get_catalog("supported-datasets/catalogs/ceda-stac-catalogue").get_collection('cmip6')
```

We have accessed the CMIP6 collection, giving us a searchable `cmip6` client object

```python
# Get the extent of the collection as a dictionary
cmip6.extent.to_dict()
```

```python
# Look for a specific item
cmip6.get_item(
    "CMIP6.ScenarioMIP.CSIRO-ARCCSS.ACCESS-CM2.ssp126.r1i1p1f1.day.pr.gn.v20210317"
)

# Another way to do the item search and return an object
cmip_item_search = client.search(
    # collections=["cmip6"],
    ids=[
        "CMIP6.ScenarioMIP.CSIRO-ARCCSS.ACCESS-CM2.ssp126.r1i1p1f1.day.pr.gn.v20210317"
    ],
    limit=10,
)
```

We can view all items from this search by iterating over the search object. In this case there is one item with a single ID. We can also pull properties from the item like the `bbox` or `assets`.

```python
for item in cmip_item_search:
    print(item.id, item.bbox)
```

### i) Plot a single timestep - CMIP6 precipitation flux

Steps:
 - Access a CMIP6 Dataset
 - Extract the dataset (Kerchunk)

```python
item_cmip = cmip_item_search[0]
product = item_cmip.get_cloud_products()
product
```

We can then perform some selections on the data (selecting the first time step), and use geoviews to plot the resulting map

```python
%matplotlib inline
# select the first time step
oneframe = ds.pr.isel(time=0)

# generate the image to show
figdata = gv.Dataset(oneframe, ["lon", "lat"], crs=crs.PlateCarree())
images = figdata.to(gv.Image)

# Plot
images.opts(cmap="viridis", colorbar=True, width=600, height=500) * gf.coastline
```

### ii) Plot an NDVI Analysis - Sentinel 2 ARD (NIR/IR)

Steps:
 - Access a Sentinel 2 ARD Dataset.
 - Open the dataset (COG) file.
 - Perform calculation for NDVI analysis over an AOI.
 - Plot resulting image.

First we define a simple coordinate converter from more familiar wgs84 lat/lon, to epsg27700 values which the xarray dataset presents with. Other options, like using an imported converter or performing a transformation using xarray are available

```python
# Coordinate Converter

def get_epsg27700_lat(wgs84_lat):
    return (wgs84_lat-49.75)*(117310/11.26) + 326955

def get_epsg27700_lon(wgs84_lon):
    return (wgs84_lon+9.01)*(117310/11.02) + 209985
```

We can then access the selected search result as before, and open the cloud product which this time is a Cloud Optimised GeoTiff (COG) file.

```python
# With CEDA DataPoint Features
product = s2_item_search[0].get_cloud_products()
product
```

```python
ds = product.open_dataset()
ds
```

Define AOI Slices to apply to the xarray DataArray, uses a coordinate conversion function defined above for temporary operation.

```python
AOI_x = slice(
    get_epsg27700_lat(51.08), 
    get_epsg27700_lat(53.08)
)
# Reverse in Y
AOI_y = slice(
    get_epsg27700_lon(-1.37),
    get_epsg27700_lon(-3.37)
)
```

Then we can apply the AOI selections to the dataset and plot the resulting image, this time with a basic matplotlib image plot. (Remember to clear previous figures, otherwise you may end up with multiple colour bars, especially when using the `savefig` method.)

```python
%matplotlib inline
# Ensure all previous sections are clear.
plt.figure().clear()

fig, ax = plt.subplots()
ax.imshow(ds.sel(band=3, x=AOI_x, y=AOI_y))
```

We can then proceed to calculate the NDVI analysis

```python
AOI = ds.sel(x=AOI_x, y=AOI_y)
AOI
```

We have created an array representation for only our AOI, now we can create ones just for the specific bands we would like to assess. Xarray loads data only when it is requested for a plot, so creating multiple versions of the array doesn't actually duplicate any data in this case.

```python
NIR = AOI.sel(band=ds.long_name.index('NIR_B8'))
Red = AOI.sel(band=ds.long_name.index('Red'))
```

```python
NDVI = (NIR - Red) / (NIR + Red)
```

Finally, the calculated data array can be plotted in the same way as before

```python
%matplotlib inline
# Ensure all previous sections are clear.
plt.figure().clear()

fig, ax = plt.subplots()
ax.imshow(NDVI)
```

## c) Streaming data with Titiler

Show some titiler links and discuss how to build your own URL
Visualise

```python

# TiTiler tile service URL
tile_url = "https://staging.eodatahub.org.uk/titiler/xarray/tiles/{z}/{x}/{y}?url=https%3A%2F%2Fdap.ceda.ac.uk%2Fbadc%2Fcmip6%2Fmetadata%2Fkerchunk%2Fpipeline1%2FScenarioMIP%2FTHU%2FCIESM%2Fkr1.0%2FCMIP6_ScenarioMIP_THU_CIESM_ssp585_r1i1p1f1_Amon_rsus_gr_v20200806_kr1.0.json&variable=rsus&tileMatrixSetId=WebMercatorQuad&reference=true&rescale=17%2C100&colormap_name=rain"

# Define a folium map centered on some global location
m = folium.Map(location=[0, 0], zoom_start=2)

# Add the TiTiler layer
folium.raster_layers.TileLayer(
    tiles=tile_url,
    attr="TiTiler",
    name="CMIP6 Data",
    overlay=True
).add_to(m)

# Add a layer control
folium.LayerControl().add_to(m)

# Display the map
m
```

```python
# First TiTiler tile service URL
tile_url1 = "https://staging.eodatahub.org.uk/titiler/xarray/tiles/{z}/{x}/{y}?url=https%3A%2F%2Fdap.ceda.ac.uk%2Fbadc%2Fcmip6%2Fmetadata%2Fkerchunk%2Fpipeline1%2FScenarioMIP%2FTHU%2FCIESM%2Fkr1.0%2FCMIP6_ScenarioMIP_THU_CIESM_ssp585_r1i1p1f1_Amon_rsus_gr_v20200806_kr1.0.json&variable=rsus&tileMatrixSetId=WebMercatorQuad&reference=true&rescale=17%2C100&colormap_name=rain"

# Second TiTiler tile service URL (example with a different variable or colormap)
tile_url2 = "https://staging.eodatahub.org.uk/titiler/xarray/tiles/{z}/{x}/{y}?url=https%3A%2F%2Fdap.ceda.ac.uk%2Fbadc%2Fcmip6%2Fmetadata%2Fkerchunk%2Fpipeline1%2FScenarioMIP%2FTHU%2FCIESM%2Fkr1.0%2FCMIP6_ScenarioMIP_THU_CIESM_ssp585_r1i1p1f1_Amon_rsus_gr_v20200806_kr1.0.json&variable=rsus&tileMatrixSetId=WebMercatorQuad&reference=true&rescale=17%2C100&colormap_name=viridis"

# Define a folium map centered globally
m = folium.Map(location=[0, 0], zoom_start=2)

# Add the first TiTiler layer
folium.raster_layers.TileLayer(
    tiles=tile_url1,
    attr="TiTiler",
    name="CMIP6 Data - Rain",
    overlay=True
).add_to(m)

# Add the second TiTiler layer
folium.raster_layers.TileLayer(
    tiles=tile_url2,
    attr="TiTiler",
    name="CMIP6 Data - Viridis",
    overlay=True
).add_to(m)

# Add a layer control to toggle between them
folium.LayerControl().add_to(m)

# Display the map
m
```
